```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
#Sentence tokenization in spacy
#object oriented

doc=nlp("Hey Dr. Sentence1. Sentence2");
for sen in doc.sents:
    print(sen)
```

```python
#Word Tokenization in spacy

for sentence in doc.sents:
    for word in sentence:
        print(word)
```
