```python
import re
```

```python
chat1="here's my phone Number: 9079073202 , phone (123)-435-5432"
chat2="phone Number:9079073202";
```

```python
pat="\d{10}|\(\d{3}\)-\d{3}-\d{4}"
re.findall(pat,chat1)
```

```python
chat1=" email: abc@gmail.com"
chat2=" mail: abc_193ui@poornima.edu"
```

```python
p="[a-zA-Z0-9_]*@[a-z]+\.[a-z]+"
re.findall(p,chat1)
```

```python
re.findall(p,chat2)
```

```python
chat="Hello I have issue with my order # 90790"
```

```python
p="order[^\d]*(\d*)"
re.findall(p,chat)
```
